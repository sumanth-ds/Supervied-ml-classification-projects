### Outliers Treatment

```python
for column in data.select_dtypes(exclude='object').drop("Attrition" , axis = 1):
    print(column , data[column].skew())
    sns.boxplot(data[column])
    plt.show()
```

```python
outlier_columns = ["MonthlyIncome" ,  "CurrentProfile" ,"LastPromotion" ,"WorkExperience" , "SalaryHikelastYear" , "NumCompaniesWorked",
                  "HomeToWork" , "Involvement"]


def outlier(data, outlier_columns):
    for column in outlier_columns:
        iq1 = np.percentile(data[column] , 25 , method="linear")
        iq3 = np.percentile(data[column] , 75 , method="linear")
        iqr = iq3 - iq1
        ll = iq1- 1.5*iqr
        ul = iq3 - 1.5*iqr
        data[column][data[column] <= ll] = ll
        data[column][data[column] >= ul] = ul
        
        
outlier(data,outlier_columns)     
```

# Loading Test Dataset

```python
test_data = pd.read_csv("Test_Dataset.csv")
test_data.head()
```

```python
test_data.info()
```

```python

## ----------------filling null valuess-------------------------


def fillna(test_data):
    for column in test_data:
        if test_data.dtypes[column] == np.number :
            test_data[column].fillna(test_data[column].median() , inplace = True )
        else:
            test_data[column].fillna(test_data[column].mode()[0] , inplace = True)
    return test_data

test_data = fillna(test_data)

## ----------------Treating outliers-------------------------


outlier_columns = ["MonthlyIncome" ,  "CurrentProfile" ,"LastPromotion" ,"WorkExperience" , "SalaryHikelastYear" , "NumCompaniesWorked",
                  "HomeToWork" , "Involvement"]


def outlier(data, outlier_columns):
    for column in outlier_columns:
        iq1 = np.percentile(data[column] , 25 , method="linear")
        iq3 = np.percentile(data[column] , 75 , method="linear")
        iqr = iq3 - iq1
        ll = iq1- 1.5*iqr
        ul = iq3 - 1.5*iqr
        data[column][data[column] <= ll] = ll
        data[column][data[column] >= ul] = ul
        
        
outlier(test_data,outlier_columns)    

## ----------------droping unwanted columns-------------------------


unwanted_columns = ["EmployeeID" , "HomeToWork" , "NumCompaniesWorked" , "SalaryHikelastYear" , "WorkExperience" , "LastPromotion" ,"CurrentProfile" , "MonthlyIncome"]

test_data.drop(unwanted_columns , axis = 1 , inplace=True)

## ----------------feature engineering on Gender column-------------------------


test_data["Gender"][test_data["Gender"] == 'F'] = "Female"


## ----------------feature engineering on MaritalStatus column-------------------------


test_data["MaritalStatus"][test_data["MaritalStatus"] == 'M'] = "Married"




## ----------------OneHot Encoding-------------------------


label_encoding = test_data.select_dtypes(exclude=np.number).columns
test_data = pd.get_dummies(test_data , columns = label_encoding , drop_first=True , dtype= int)

test_data.head()
```

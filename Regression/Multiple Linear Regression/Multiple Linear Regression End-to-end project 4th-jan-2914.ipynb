```python
#we have to handle outliers for the listed variable

# Avg. Area Income , Avg. Area House Age, Avg. Area Number of Rooms ,Area Population
```

```python
Q1 = housing.quantile(0.25)
Q3 = housing.quantile(0.75)
IQR = Q3 -Q1

pos_Outlier = Q3 + 1.5*IQR
neg_Outlier = Q1 - 1.5*IQR


print("Q1 :" ,Q1)
print("*"*70)

print("Q3 :" ,Q3)
print("*"*70)



print("IQR :" ,IQR)
print("*"*70)


print("Positive quantile Range :" , pos_Outlier)
print("*"*70)



print("Negitive quantile Range :" , neg_Outlier)
print("*"*70)

```

```python
def caping(col):
    global housing
    col_q1 = housing[col].quantile(0.25)
    col_q3 = housing[col].quantile(0.75)
    iqr = col_q3 - col_q1
    lower = col_q1 - 1.5*iqr
    upper = col_q3 + 1.5*iqr 
    
    housing[col] = np.where(housing[col]>upper,upper,np.where(housing[col]<lower,lower,housing[col]))
    

    
caping_list = ['Avg. Area Income' ,'Avg. Area House Age', 'Avg. Area Number of Rooms' ,'Area Population']
for i in caping_list:
    caping(i)
```
